```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv('train.csv')
```

```python
test = pd.read_csv('test.csv')
```

```python
train = train.drop(['PassengerId', 'Name', 'Ticket'], axis='columns')
test = test.drop([ 'Name', 'Ticket'], axis='columns')
```

```python
train = train.drop(['Cabin'], axis=1)
test = test.drop(['Cabin'], axis=1)
```

```python
d = pd.get_dummies(train['Sex'], drop_first=True)
d
```

```python
train['Sex'] = d
train
```

```python
d = pd.get_dummies(test['Sex'], drop_first=True)
d
```

```python
test['Sex'] = d
```

```python
test
```

```python
train = train.dropna(subset=['Embarked'])
```

```python
train.shape
```

```python
test.isna().sum()
```

```python
test.Fare = test.Fare.fillna(test.Fare.mean())
```

```python
test.isna().sum()
```

```python
e = pd.get_dummies(train.Embarked, drop_first=True)
train = pd.concat([train, e], axis=1)
train
```

```python
train = train.drop('Embarked', axis=1)
```

```python
e = pd.get_dummies(test.Embarked, drop_first=True)
test = pd.concat([test, e], axis=1)
test = test.drop('Embarked', axis=1)
test
```

```python
train.groupby('Pclass').mean()
```

```python
def fill_age(cols):
    Age, Pclass = cols
    if not pd.isnull(Age):
        return Age
    else:
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        else:
            return 25
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(fill_age, axis=1)
test['Age'] = test[['Age', 'Pclass']].apply(fill_age, axis=1)
```

```python
test.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder, LabelBinarizer

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
X_train = train.drop('Survived', axis=1)
X_train
```

```python
y_train = train.Survived
```

```python
X_test = test.drop('PassengerId', axis=1)
X_test
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
pred = pd.Series(pred)
pred
```

```python
final = pd.concat([test.PassengerId, pred], axis=1)
final.head(15)
```

```python
final.columns = ['PassengerId', 'Survived']
```

```python
final.to_csv('final.csv', index=False)
```
